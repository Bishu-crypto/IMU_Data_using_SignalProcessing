# Phone IMU — Spectral Analysis
Welch PSD and jitter analysis for SP-04 project.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Load your logged IMU data here
# data = np.load('imu_log.npy')
# ax = data[:, 0]

# Example: generate synthetic accelerometer data
fs = 100.0
t  = np.arange(0, 30, 1/fs)
ax = 0.1*np.sin(2*np.pi*1.5*t) + 0.01*np.random.randn(len(t))

f, psd = welch(ax, fs=fs, window='hann', nperseg=256, noverlap=128)
noise_floor = 10*np.log10(np.median(psd))
print(f'Noise floor: {noise_floor:.1f} dB/Hz')

plt.figure(figsize=(10, 4))
plt.semilogy(f, psd)
plt.xlabel('Frequency (Hz)')
plt.ylabel('PSD (m/s²)²/Hz')
plt.title('Accelerometer Power Spectral Density (Welch)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('psd_plot.png', dpi=150)
plt.show()
```
